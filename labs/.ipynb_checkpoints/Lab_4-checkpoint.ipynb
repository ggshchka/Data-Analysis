### 4. Исследование корреляционной связи


- Случайная величина $U = \lambda X + (1-\lambda)Y$
- Случайная величина $V = \lambda X^3 + (1-\lambda)Y^3$
- $\lambda \in [0;1]$

```python
lambda_val = np.linspace(0, 1, 100)
def u(l, x, y):
    return l * x + (1 - l) * y

def v(l, x, y):
    return l * x**3 + (1 - l) * y**3
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.set_title('Графики зависимостей $КК_{XU}$ от $\lambda$')
ax.plot(lambda_val, [stats.pearsonr(x, u(l, x, y))[0] for l in lambda_val], label='$r_{XV}$')
ax.plot(lambda_val, [stats.spearmanr(x, u(l, x, y))[0] for l in lambda_val], label='$r_{XV}$')
ax.plot(lambda_val, [stats.kendalltau(x, u(l-0.0001, x, y))[0] for l in lambda_val], label='$r_{XV}$')
ax.legend();
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.set_title('Графики зависимостей $КК_{XV}$ от $\lambda$')
ax.plot(lambda_val, [stats.pearsonr(x, v(l, x, y))[0] for l in lambda_val], label='$r_{XV}$')
ax.plot(lambda_val, [stats.spearmanr(x, v(l, x, y))[0] for l in lambda_val], label='$r_{XV}$')
ax.plot(lambda_val, [stats.kendalltau(x, v(l-0.0001, x, y))[0] for l in lambda_val], label='$r_{XV}$')
ax.legend()
```

Выводы:
- При увелечении λ увеличивается влияние случайной величины X на величину V и уменьшается влияние Y , поэтому, как видно на графиках коэффициенты корреляции растут.
- При увелечении λ увеличивается влияние случайной величины X на величину V и уменьшается влияние Y , поэтому, как видно на графиках коэффициенты корреляции растут.

```python
v0 = v(0, x, y)
v1 = v(1, x, y)
fig, ax = plt.subplots(2, 2, figsize=(15,15))
ax[0][0].set_title('$X, V$ при $\\lambda = 0$')
ax[0][0].scatter(x, v0)

ax[0][1].set_title('Ранги $X, V$ при $\\lambda = 0$')
ax[0][1].scatter(stats.rankdata(x), stats.rankdata(v0))


ax[1][0].set_title('$X, V$ при $\\lambda = 1$')
ax[1][0].scatter(x, v1)

ax[1][1].set_title('Ранги $X, V$ при $\\lambda = 1$')
ax[1][1].scatter(stats.rankdata(x), stats.rankdata(v1))
```

Выводы:
- При λ = 0 мы видим, отстутсвие какой либо зависимости V(X)
- При λ = 1 можно сделать вывод, что V зависит от X функционально.
